```python
df
```

```python
df_encoded = pd.get_dummies(df, columns = ['vote', 'gender'], drop_first = True) 
df_encoded
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled = scaler.fit_transform(df_encoded) 
scaled
```

```python
scaled_df = pd.DataFrame(scaled, columns=df_encoded.columns)
```

```python
scaled_df
```

```python
scaled_df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = scaled_df.drop("vote_Labour",axis = 1)
y = scaled_df['vote_Labour']
```

```python
y.shape
```

```python
x.shape
```

```python
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size = .30 ,random_state = 1)
```

```python
print('x_train', x_train.shape)

print('y_train', y_train.shape)

print('x_test', x_test.shape)

print('y_test', y_test.shape)
```

```python
#Logistic Regression and LDA 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(tol=0.0001,max_iter=2000,random_state= 0)
```

```python
lr.fit(x_train,y_train)
```

```python
y_train_predict=lr.predict(x_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_train ,y_train_predict))
```

```python
print(sns.heatmap(confusion_matrix(y_train,y_train_predict),annot=True, fmt="d", cmap=mycolors))
plt.show()
confusion_matrix(y_train,y_train_predict)
```

```python
print(classification_report(y_test,lr.predict(x_test)))
```

```python
print(sns.heatmap(confusion_matrix(y_test,lr.predict(x_test)),annot=True, fmt="d", cmap=mycolors))
confusion_matrix(y_test,lr.predict(x_test))
```

```python
from sklearn.metrics import mean_squared_error
print('Error for Train Data:', mean_squared_error(y_train_predict, y_train))
print('Error for Test Data:', mean_squared_error(lr.predict(x_test), y_test))
```

```python
print('Error for Test Data:', mean_squared_error(lr.predict(x_test), y_test))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python

probs = lr.predict_proba(x_train)
probs = probs[:, 1]
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_train, probs)
print('AUC: %.3f' % auc)
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train, probs)
plt.plot([0, 1], [0, 1], linestyle='solid')
plt.plot(fpr, tpr, marker='.')
plt.show()
```
